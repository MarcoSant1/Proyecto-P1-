```python
#Espectros de Carga Camión C2
def plot_distribution(C2, columns):
    for col in columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(C2[col], kde=True)
        plt.title(f'Espectro de carga en el "{col}" del Camión Tipo C2')
        plt.xlabel('Toneladas')
        plt.ylabel('Cantidad de Vehículos')
        plt.show()
# Visualizar la distribución de algunas variables
columns_to_plot_C2 = ['EJE SENCILLO (Ton)', 'EJE SENCILLO DUAL (Ton)']
plot_distribution(C2, columns_to_plot_C2)
```

## **4. Conclusiones y recomendaciones basadas en los hallazgos del análisis**

> Los vehículos con PBV superiores a 40 toneladas representan una amenaza crítica para la durabilidad de la infraestructura vial. La alta carga sobre los ejes, especialmente en vehículos tipo T3S2, T3S3 y T3S2R4, puede acelerar el deterioro de las carreteras y puentes, resultando en un aumento de los costos de mantenimiento y reparación.<br> <br>

> La distribución desigual del peso en los ejes de los vehículos indica que ciertos ejes están soportando cargas excesivas. Esta situación puede llevar a un desgaste prematuro y desigual de las carreteras, aumentando el riesgo de daños estructurales y la necesidad de reparaciones frecuentes. La infraestructura vial actual podría no estar preparada para soportar estas cargas extremas, lo que pone en riesgo la seguridad y eficiencia del transporte..<br> <br>


> El análisis revela una diversidad extrema en las configuraciones de carga vehicular, con pesos brutos vehiculares (PBV) que varían desde 2.862 kg hasta 99.235 kg. La falta de uniformidad en las configuraciones de carga presenta desafíos significativos para la planificación y el mantenimiento de la infraestructura vial.<br> <br>

> Se recomienda implementar sistemas de pesaje dinámico: Usar tecnología avanzada de pesaje dinámico en las carreteras para monitorear continuamente los pesos de los vehículos en movimiento. Esto permitirá una recopilación de datos más precisa y en tiempo real.<br> <br>

> Además, se aconseja la presencia y la frecuencia de estaciones de pesaje en puntos estratégicos para asegurar que los vehículos cumplan con los límites legales de peso. Al igual que implementar sanciones más estrictas para los vehículos que excedan los límites permitidos y realizar controles aleatorios para disuadir el incumplimiento.<br> <br>
